```python
import numpy as np

# 初始化矩阵大小为 17x25
matrix = np.zeros((17, 25))

# 设置最大值的位置
max_val = 0.6910
matrix[15, 5] = max_val

# 设置x坐标区间
x_intervals = {
    (0, 6): [50.21, 58.08],
    (6, 13): [55.11, 61.98],
    (13, 17): [60.11, 69.08]
}

# 用于存储所有已使用的数值，确保不重复
used_values = set()

# 填充矩阵的每一列
for x in range(25):
    if x < 6:
        min_val, max_val = x_intervals[(0, 6)]
    elif x < 13:
        min_val, max_val = x_intervals[(6, 13)]
    else:
        min_val, max_val = x_intervals[(13, 17)]
    
    # 对y坐标的影响进行处理
    for y in range(17):
        if y < 6:
            # y<6时数据稳定上升，使用线性增加
            value = min_val + (max_val - min_val) * (y / 6)
        elif y == 6:
            # y=6时是最大值
            value = max_val
        else:
            # y>6时数据不稳定地波动下降
            # 增加更大的波动范围（增大波动的幅度）
            fluctuation = np.random.uniform(-0.1, 0.1)  # 更大的波动
            value = max_val - (y - 6) * (max_val - min_val) / 10 + fluctuation
        
        # 保证没有重复的数值
        while value in used_values:
            fluctuation = np.random.uniform(-0.1, 0.1)  # 重新生成一个波动
            value = max_val - (y - 6) * (max_val - min_val) / 10 + fluctuation
        
        matrix[y, x] = value
        used_values.add(value)

# 输出矩阵
matrix
```
